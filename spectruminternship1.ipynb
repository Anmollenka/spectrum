```python
#IMPORT NUMPY AND MATPLOTLIB
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#GIVEN IS THE DICTIONARY CONTAINING THE SCORES OBTAINED EACH DAY
scores = {"Day 1": 100, "Day 2": 108, "Day 3":112, "Day 4":115, "Day 5":150,
          "Day 6":178, "Day 7": 143, "Day 8": 132, "Day 9":190, "Day 10": 235,
          "Day 11":253, "Day 12": 298, "Day 13": 328, "Day 14":390, "Day 15": 257,
          "Day 16":288, "Day 17": 393, "Day 18": 425, "Day 19":458, "Day 20": 450,
          "Day 21":473, "Day 22": 333, "Day 23": 452, "Day 24":490, "Day 25": 495,
          "Day 26":488, "Day 27": 543, "Day 28": 532, "Day 29":590, "Day 30": 605}

```

```python
Scores=np.array(list(scores.values()))
```

```python
Days=np.arange(1,31)
```

```python
#PLOTTING A GRAPH BETWEEN SCORES AND DAYS
fig=plt.figure()
ax=fig.add_axes([0,0,2,2])
ax.set_title("SCORE VS. DAYS")
ax.set_xlabel("SCORE")
ax.set_ylabel("DAYS")
plt.plot(Scores,Days,color="purple",lw=2,marker='.',markersize=12,markerfacecolor="black")
```

```python
print("MEDIAN SCORE :    "+str(np.median(Scores)))
print("MEAN   SCORE :    "+str(np.mean(Scores)))
print("MINIMUM SCORE:    "+str(np.min(Scores)))
print("MAXIMUM SCORE:    "+str(np.max(Scores)))
```
